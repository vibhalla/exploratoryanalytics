## An example of supervised learning - using Decision Tree

##### # Dataset used from https://archive.ics.uci.edu/ml/datasets/Website+Phishing

```python


# Import the required packages
import pandas as pd
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plot
import seaborn as sn
from sklearn.metrics import classification_report, confusion_matrix,mean_absolute_error,mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
# Load the data from the CSV file
phishData = pd.read_csv("../Data/PhishingData.csv")
phishFeatures = ['SFH', 'popUpWidnow', 'SSLfinal_State', 'Request_URL', 'URL_of_Anchor', 'web_traffic', 'URL_Length', 'age_of_domain', 'having_IP_Address']
phishClass = ['-1','0','1'] 
print("Header :\n" , phishData.columns , "\n")
print("\n\nDescription of the data : \n")
phishData.describe()
```

```python
# Preparing the train and test data
Features = phishData.iloc[: , :9].values
Target = phishData.iloc[: , -1].values
xTrain, xTest, yTrain, yTest = train_test_split(Features, Target, test_size = .2, random_state = 42)
```

```python
# Train and Test

decision_tree = tree.DecisionTreeClassifier(min_samples_leaf=20, max_depth=6)


decision_tree = decision_tree.fit(xTrain, yTrain)
yPred = decision_tree.predict(xTest)

print("\n Classification Report: \n")
print(classification_report(yTest, yPred))

print("\nConfusion Matrix: \n")
cm = pd.DataFrame(confusion_matrix(yTest, yPred), columns=['Predicted -1', 'Predicted 0', 'Predicted 1'],
    index=['Actual -1', 'Actual 0', 'Actual 1'])
print(cm)
sn.heatmap(cm, annot=True)

tree.export_graphviz(decision_tree, out_file = 'phishing.dot', feature_names = phishFeatures,
                     class_names = phishClass, filled=True, rounded=True) 
```

As we can see from the Confusion Matrix that out of 271 records 218 records were correctly predicted
which gives the setup an accuracy of around 80%. Same can be validated using the classification report.

## An example of unsupervised learning - using K-Means clustering and Gaussian Mixture

##### Dataset used - http://ggobi.org/book/data/olive.csv
##### The data contains the percentage of fatty acid in composition of Olive Oil and is used to check the authenticity of the olive oil.

```python
# Import the required packages
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cross_validation import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the CSV file and load the data

oliveData = pd.read_csv("../Data/olive.csv")
print("Raw Header :\n" , oliveData.columns , "\n")
print("\n\nDescription of the data : \n")
oliveData.describe()
```

```python
# We drop the unwanted column - Area and also the target column Region as it is not needed for unsupervised learning
# Then e prep the data. Standardize the data and then make the data 2 dimensional 
# using Principal Components Analysis. Then we create the Train and Test sets

features = ['palmitic', 'palmitoleic', 'stearic', 'oleic', 'linoleic', 'linolenic','arachidic', 'eicosenoic']

# Separating out the features
x = oliveData.loc[:, features].values

# Extract the target
y = oliveData.loc[:,['region']].values

x = StandardScaler().fit_transform(x)

pca = PCA(n_components=2)
x = pca.fit_transform(x)

explained_variance = pca.explained_variance_ratio_
print("Explained Variance: ", explained_variance , "\n\n")

oliveDF = pd.DataFrame(data = x , columns = ['pc1', 'pc2'])
print(oliveDF.head())

X_Train, X_Test = train_test_split( oliveDF, test_size=0.20, random_state=45)
print("\n\nTrain size :",len(X_Train),"\nTest Size :",len(X_Test))
```

```python
plt.scatter(X_Test.iloc[:,0],X_Test.iloc[:,1],c='red')
plt.title('Before Clustering')
plt.show()
```

```python
centroidList = []
numClusters = 4
sample = X_Test.sample(n=numClusters)
for i in range(numClusters):
    centroid = []
    for d in range(len(X_Test.columns.values)):
        centroid.append(sample.iloc[i, d])
    centroidList.append(centroid)
print("Centroid List: ",centroidList)
X_Test.plot.scatter(x='pc1',y='pc2',c='cyan')
for ix, centroid in enumerate(centroidList):
    plt.plot(centroid[0], centroid[1], color='red', marker='*', ms=10)
plt.title("Data Before Clustering")
plt.xlabel("pc1")
plt.ylabel("pc2")
plt.show()
```

```python
oldCent = []
numIter = 0
numCentSel = 0
centDiff = 100000
#iterate until the centroids are stable
while centDiff > .001:
    numIter = numIter + 1
    centGrp = [[] for centroid in range(numClusters)]
    #iterate over the data
    for i, x in X_Test.iterrows():
        dist = 0
        min_dist = 100000
        numCentSel = 0
        for centNum in range(len(centroidList)):
            distance = 0
            for sampleNum in range(len(X_Test.columns.values)):
                distance = distance + (centroidList[centNum][sampleNum] - x[sampleNum])**2
                if distance < min_dist:
                    min_dist = distance
                    numCentSel = centNum
        centGrp[numCentSel].append(list(x))
        
    for i, x in enumerate(centGrp):
        if i == 0:
            color = 'yellow'
            marker = '.'
        elif i == 1:
            color = 'cyan'
            marker = '.'
        elif i == 2:
            color = 'blue'
            marker = '.'
        else:
            color = 'green'
            marker = '.'
        for sample in x:
            plt.plot(sample[0], sample[1], color=color, marker=marker)
    for i, centroid in enumerate(centroidList):
        plt.plot(centroid[0], centroid[1], color='red', marker='*', ms=10)
    plt.title('K-Means: Data after iteration :')
    plt.xlabel("pc1")
    plt.ylabel("pc2")
    plt.show()

#preserving old centroid values
    oldCent = centroidList
    centroidList = []
    for cent in centGrp:
        sumCent = [sum(x) for x in zip(*cent)]
        avgCent = [y/len(cent) for y in sumCent]
        centroidList.append(avgCent)
    
    centDiff = 0
    centFeat = len(centroidList[0])
    for feature in range(centFeat):
        for cent, prevCent in zip(centroidList, oldCent):
            centDiff = centDiff + abs(cent[feature] - prevCent[feature])
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X_Train)
centroids = kmeans.cluster_centers_
labels = kmeans.predict(X_Test)
plt.scatter(X_Test.iloc[:,0],X_Test.iloc[:,1],c=labels)
for idx, centroid in enumerate(centroids):
    plt.plot(centroid[0], centroid[1], color='red', marker='*', ms=10)
plt.title('Using Kmeans')
plt.show()
```

```python
#Gaussian Mixture
GMM = GaussianMixture(n_components=4).fit(X_Train)
labels = GMM.predict(X_Test)
converged = GMM.converged_
print('Converged:', converged) # To validate if the data has converged
plt.scatter(X_Test.iloc[:,0],X_Test.iloc[:,1],c=labels)
plt.title('Using Gaussian Mixture')
plt.show()
```

As can be seen above both KMeans and GaussianMixture algorithm result into 4 clusters similar to the manual code
