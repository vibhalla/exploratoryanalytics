## An example of supervised learning - using Decision Tree

##### Dataset used from https://archive.ics.uci.edu/ml/datasets/Website+Phishing

```python
# Import the required packages
import pandas as pd
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plot
import seaborn as sn
from sklearn.metrics import classification_report, confusion_matrix,mean_absolute_error,mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
# Load the data from the CSV file
phishData = pd.read_csv("../Data/PhishingData.csv")
phishFeatures = ['SFH', 'popUpWidnow', 'SSLfinal_State', 'Request_URL', 'URL_of_Anchor', 'web_traffic', 'URL_Length', 'age_of_domain', 'having_IP_Address']
phishClass = ['-1','0','1'] 
print("Header :\n" , phishData.columns , "\n")
print("\n\nDescription of the data : \n")
phishData.describe()
```

```python
# Preparing the train and test data
Features = phishData.iloc[: , :9].values
Target = phishData.iloc[: , -1].values
xTrain, xTest, yTrain, yTest = train_test_split(Features, Target, test_size = .2, random_state = 42)
```

```python
# Train and Test

decision_tree = tree.DecisionTreeClassifier(min_samples_leaf=20, max_depth=6)


decision_tree = decision_tree.fit(xTrain, yTrain)
yPred = decision_tree.predict(xTest)

print("\n Classification Report: \n")
print(classification_report(yTest, yPred))

print("\nConfusion Matrix: \n")
cm = pd.DataFrame(confusion_matrix(yTest, yPred), columns=['Predicted -1', 'Predicted 0', 'Predicted 1'],
    index=['Actual -1', 'Actual 0', 'Actual 1'])
print(cm)
sn.heatmap(cm, annot=True)

tree.export_graphviz(decision_tree, out_file = 'phishing.dot', feature_names = phishFeatures,
                     class_names = phishClass, filled=True, rounded=True) 
```

As we can see from the Confusion Matrix that out of 271 records 218 records were correctly predicted
which gives the setup an accuracy of around 80%. Same can be validated using the classification report.
